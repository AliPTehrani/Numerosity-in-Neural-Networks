Function to plot the different ROIs

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
def plot_avg_nc():
  models = ["Alexnet" ,"Alexnet" ,"Alexnet" , "VGG-11" ,"VGG-11" ,"VGG-11" ,"VGG-19" ,"VGG-19" ,"VGG-19" ,
            "ResNet-34","ResNet-34","ResNet-34","ResNet-152","ResNet-152","ResNet-152", 
             "CorNet-Z","CorNet-Z","CorNet-Z", "CorNet-R","CorNet-R","CorNet-R","CorNet-S","CorNet-S","CorNet-S"]

  values = [
            alexnet["V13"] , alexnet["V3ABV7"],alexnet["IPS15"],
            vgg11["V13"], vgg11["V3ABV7"],vgg11["IPS15"],
            vgg19["V13"], vgg19["V3ABV7"],vgg19["IPS15"],
            resnet34["V13"],resnet34["V3ABV7"],resnet34["IPS15"],
            resnet152["V13"],resnet152["V3ABV7"],resnet152["IPS15"],
            cornetZ["V13"],cornetZ["V3ABV7"],cornetZ["IPS15"],
            cornetR["V13"],cornetR["V3ABV7"],cornetR["IPS15"],
            cornetS["V13"],cornetS["V3ABV7"],cornetS["IPS15"],
           ]

  ROI = ["V13","V3ABV7","IPS15","V13","V3ABV7","IPS15","V13","V3ABV7","IPS15","V13","V3ABV7","IPS15",
         "V13","V3ABV7","IPS15","V13","V3ABV7","IPS15","V13","V3ABV7","IPS15","V13","V3ABV7","IPS15"]

  #values_V3ABV7 = [alexnet["V3ABV7"] , vgg19["V3ABV7"], resnet152["V3ABV7"],cornetS["V3ABV7"]]
  #values_IPS15 = [alexnet["IPS15"] , vgg19["IPS15"], resnet152["IPS15"],cornetS["IPS15"]]
  df = pd.DataFrame({ "values" : values, "ROI" : ROI,"models" : models})

  sns.set_style('darkgrid')
  s = sns.barplot(x = "models", y = "values", hue = "ROI",data = df)
  plt.title("Average noise ceilling percentage")
  plt.ylabel("Average percantege of noise ceilling")
  plt.xlabel("Models")
  plt.xticks(rotation=45)
  plt.ylim(0,100)
  figure(figsize=(10, 8), dpi=80)
  plt.show()
  plt.savefig("/content/sample_data/noise_ceilling_percentage.png")
  plt.close()
  
```

```python
plot_avg_nc()
```
